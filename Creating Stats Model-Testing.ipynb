```python
import pandas as pd
import statsmodels.api as sm
```

```python
file_path = r'D:/Documents/GitHub/Statistical-Modelling-Project/filtered_merged_data.csv'

# Read the CSV file into a DataFrame
merged_df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify import
print(merged_df.head())
```

```python
features = ['Latitude', 'Longitude', 'Rating', 'POI Count']
target = 'Free_Bikes'  # Example target variable, replace with your target

# Prepare X (features) and y (target)
X = merged_df[features]
y = merged_df[target]

# Add constant term to X (for intercept)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

```python
import pandas as pd
import statsmodels.api as sm

# Assuming merged_df contains your data with relevant columns

# Define features and target
features = ['Latitude', 'Longitude', 'Weighted_Rating', 'POI Count']  # Adjusted to include Weighted_Rating
target = 'Free_Bikes'  # Example target variable, replace with your target

# Calculate Weighted Rating based on Free_Bikes (example: simple ratio)
merged_df['Weighted_Rating'] = merged_df['Rating'] * (merged_df['Free_Bikes'] / merged_df['Free_Bikes'].max())

# Prepare X (features) and y (target)
X = merged_df[features]
y = merged_df[target]

# Add constant term to X (for intercept)
X = sm.add_constant(X)

# Fit the linear regression model
model_weighted_rating = sm.OLS(y, X).fit()

# Print the model summary
print(model_weighted_rating.summary())
```
